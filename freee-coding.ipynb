```python
A =[0,3,5,2,3,4,3,4,2,2,3,6,2,0]
B =[0,4,3,3,2,3,0]
def A_UP_1(time,position):
    arrive_time =sum(A[0:position])
    which_7or_13 =0 #奇数であれば7行、偶数であれば13行
    if time>=24 or(time>=0 and time<=4):
        return "finish",which_7or_13
    if time ==5 and 55+arrive_time>=60:
        return  "finish",which_7or_13
    a_up_ans_1 =[]
    search_hour =5
    minue =arrive_time+55
    if minue >=60:
        search_hour +=1
        minue -=60
    
    while time !=search_hour:
        minue +=5
        which_7or_13 +=1
        if minue>=60:
            minue -=60
            search_hour +=1
    if time ==23:
        search =minue-arrive_time #(到着する時間)-(到着するのにかかる時間)がマイナスであればA1を出発したのは22時台
        while search <0:
            a_up_ans_1.append(minue)
            search +=5
            minue +=5
            which_7or_13 +=1
        return a_up_ans_1,which_7or_13  
    
    while minue<60:
        a_up_ans_1.append(minue)
        minue +=5
        which_7or_13 +=1
    return a_up_ans_1,which_7or_13


#A7発A1行の電車は1日を通じてA1発A7行と同じであるため103本出発する
def A_DOWN_7(time,position):
    cnt_start =0 #A7発A1行のカウント
    arrive_time =sum(A[position:7])
    if time>=24 or(time>=0 and time<=5):
        return "finish"
    a_down_ans_7 =[]
    search_hour =6
    minue =arrive_time+6
    while time !=search_hour:
        minue +=10
        cnt_start +=1
        if minue>=60:
            minue -=60
            search_hour +=1
  
    while minue<60 and cnt_start <103 :
        a_down_ans_7.append(minue)
        minue +=10
        cnt_start +=1
   
    return a_down_ans_7

    



def B_UP(time,position):
    if time>=24 or(time>=0 and time<=4):
        return "finish"

    arrive_time =sum(B[0:position])
    b_up_ans =[]
    search_hour =6
    minue =arrive_time
    while time !=search_hour:
        minue +=6
        if minue>=60:
            minue -=60
            search_hour +=1
    if time ==23:
        search =minue-arrive_time #(到着する時間)-(到着するのにかかる時間)がマイナスであればB1を出発したのは22時台
        while search <0:
            b_up_ans.append(minue)
            search +=6
            minue +=6
        return b_up_ans 
    
        while minue>0:
             b_up_ans.append(minue)
             minue -=6
        return b_up_ans
    
    while minue<60:
        b_up_ans.append(minue)
        minue +=6
   
    return b_up_ans
    
#B1発A7行は１日あたり170本のが発車するためA7発B1行も170本
def B_DOWN(time,position):
    if time>=24 or(time>=0 and time<=4):
        return "finish"

    cnt_start =0 #A7発B1行の本数を数える
    arrive_time =sum(B[position:7])
    b_down_ans =[]
    search_hour =6
    minue =arrive_time+11
    while time !=search_hour:
        minue +=6
        cnt_start +=1
        if minue>=60:
            minue -=60
            search_hour +=1
   
   
    while minue<60 and cnt_start <170:
        b_down_ans.append(minue)
        cnt_start +=1
        minue +=6

  
    return b_down_ans
    


def A_DOWN_13(time,position):
    arrive_time =sum(A[position:-1])
    if time ==5 and 52+arrive_time>=60:
        return  "finish"
    if time>=24 or(time>=0 and time<=4):
        return "finish"
    a_down_ans_13 =[]
    search_hour =5
    minue =arrive_time+52
    if minue >=60:
        minue -=60
        search_hour +=1

    while time !=search_hour:
        minue +=10
        if minue>=60:
            minue -=60
            search_hour +=1
    if time ==23:
        search =minue-arrive_time
        while search <0:
            a_down_ans_13.append(minue)
            search +=10
            minue +=10
        return a_down_ans_13 
    while minue<60:
        a_down_ans_13.append(minue)
        minue +=10
   
    return a_down_ans_13




def A1_goto_A13(time,position):
    ans_array,which_7or_13 =A_UP_1(time,position)
    go_to_13 =[]
    #23時台の最後の電車は必ずA7行となるため特殊な処理をする
    if time ==23:
        which_7or_13 -=1
        #A1から出発する23時台の最後から2番目が13行
        if which_7or_13%2==0:
            ans_index =-2
        #A1から出発する23時台の最後から2番目がA7行
        else:
            ans_index =-3
    
    else:
        #A1から出発する指定時間の最後の電車がA13行
        if  which_7or_13%2==0:
            ans_index =-1
        #A1から出発する指定時間の最後の電車がA7行
        else:
            ans_index =-2
            
    while True:
        if  len(ans_array)+ ans_index<0:
            return go_to_13
        go_to_13.insert(0,ans_array[ans_index])
        ans_index -=2




def A7_goto_A13(position):
    arrive_time =sum(A[7:position])+10
    return arrive_time



def ans_output(time_str,ans_array):
    print(time_str,end="")
    print(":",end=" ")
    for i in range(len(ans_array)):
        if int(ans_array[i])<10:
            ans =str("0")+str(ans_array[i])
        else:
            ans =str(ans_array[i])
        print(ans,end=" ")




def ans(line_ab,position,up_down,time):
    if line_ab=="A" and position<=6 and up_down =="U":
        ans_array,damy =A_UP_1(time,position)
        if   len(ans_array) ==0 :
            print("No train")
        elif  ans_array[0] =="f":
            print("No train")
        else:
            ans_output(time_str,ans_array)  
   

    if line_ab=="A" and position>=7 and up_down =="U":
        ans_array,which_7or_13 =A_UP_1(time,position)
        if ans_array =="finish" or len(ans_array) ==0:
            print("No train")
        else:
            ans_array =A1_goto_A13(time,position)
            #6時10分にA7から出発する電車
            if time ==6:
                ans_array.append(A7_goto_A13(position))
                ans_array.sort()
            ans_output(time_str,ans_array)


    if line_ab=="A"  and up_down =="D" and position<=7:
        ans_array =A_DOWN_7(time,position)
        ans_array2 =A_DOWN_13(time,position)
        #A13発の最終はA7行
        if time ==23:
            del ans_array2[-1] 
        ans_array +=ans_array2
        ans_array =list(set(ans_array))
        ans_array.sort()
       
        if   len(ans_array) ==0 :
            print("No train")
        elif  ans_array[0] =="f":
            print("No train")
    
        else:
            ans_output(time_str,ans_array)
        

    if line_ab=="A"  and up_down =="D" and position>=8:
        ans_array =A_DOWN_13(time,position)
       
        if   len(ans_array) ==0 :
            print("No train")
        elif  ans_array[0] =="f":
            print("No train")
    
        else:
            ans_output(time_str,ans_array)
                

    if line_ab=="B" and up_down =="U":
        ans_array =B_UP(time,position)
        if   len(ans_array) ==0 :
            print("No train")
        elif  ans_array[0] =="f":
            print("No train")
    
        else:
            ans_output(time_str,ans_array)

    
    if line_ab=="B" and up_down =="D":
        ans_array =B_DOWN(time,position)
        if   len(ans_array) ==0 :
            print("No train")
        elif  ans_array[0] =="f":
            print("No train")
        else:
            ans_output(time_str,ans_array)

line_ab,position,up_down,time =input().split()
if len(position)==3:
    position =int(position[1] + position[2])
else:
    position =int(position[1])

time_str =time
time =int(time)
ans(line_ab,position,up_down,time)

```
